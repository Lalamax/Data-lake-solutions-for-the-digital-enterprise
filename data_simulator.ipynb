```python
from faker import Faker
import random
import pandas as pd
from datetime import datetime, timedelta

fake = Faker()

# données fictives de transactions
data = []
for _ in range(1000): 
    client_id = fake.uuid4()
    client_name = fake.name()
    product = fake.word()
    transaction_date = fake.date_this_year()
    amount = round(random.uniform(10, 500), 2)  
    payment_method = random.choice(['Credit Card', 'PayPal', 'Bank Transfer'])
    
    data.append([client_id, client_name, transaction_date, product, amount, payment_method])


df_transactions = pd.DataFrame(data, columns=['ClientID', 'ClientName', 'TransactionDate', 'Product', 'Amount', 'PaymentMethod'])


df_transactions.to_csv('transactions.csv', index=False)
```

```python
import random
from datetime import datetime, timedelta
import pandas as pd

# logs de serveurs
logs = []
for _ in range(1000):  
    ip_address = f"{random.randint(100, 255)}.{random.randint(100, 255)}.{random.randint(100, 255)}.{random.randint(100, 255)}"
    url = random.choice(["/home", "/about", "/product", "/contact"])
    status_code = random.choice([200, 404, 500])
    timestamp = datetime.now() - timedelta(days=random.randint(0, 30))  
    logs.append([ip_address, url, status_code, timestamp])


df_logs = pd.DataFrame(logs, columns=['IP_Address', 'URL', 'Status_Code', 'Timestamp'])

df_logs.to_csv('web_logs.csv', index=False)
```

```python
import random
from faker import Faker
import json

fake = Faker()

# posts sur les réseaux sociaux
social_data = []
for _ in range(1000):  
    user = fake.user_name()
    message = fake.sentence(nb_words=10)
    post_date = fake.date_this_year().strftime('%Y-%m-%d')  
    likes = random.randint(0, 1000)
    retweets = random.randint(0, 500)
    
    post = {
        'user': user,
        'message': message,
        'post_date': post_date,
        'likes': likes,
        'retweets': retweets
    }
    
    social_data.append(post)


with open('social_data.json', 'w') as json_file:
    json.dump(social_data, json_file, indent=4)
```

```python
import random
from datetime import datetime, timedelta
import json
from faker import Faker

fake = Faker()

# données de la campagne publicitaire
ad_campaigns = []
for _ in range(100): 
    campaign_id = fake.uuid4()
    start_date = datetime.now() - timedelta(days=random.randint(1, 30))
    end_date = start_date + timedelta(days=random.randint(1, 30))
    budget = round(random.uniform(1000, 10000), 2)
    clicks = random.randint(100, 1000)
    impressions = random.randint(1000, 10000)
    ctr = round(clicks / impressions, 4) if impressions > 0 else 0
    
    ad_campaign = {
        'campaign_id': campaign_id,
        'start_date': start_date.strftime('%Y-%m-%d %H:%M:%S'),  
        'end_date': end_date.strftime('%Y-%m-%d %H:%M:%S'),  
        'budget': budget,
        'clicks': clicks,
        'impressions': impressions,
        'ctr': ctr
    }
    
    ad_campaigns.append(ad_campaign)


with open('ad_campaigns.json', 'w') as json_file:
    json.dump(ad_campaigns, json_file, indent=4)
```
